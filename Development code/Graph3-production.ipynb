```python
import pandas as pd
import numpy as np
```

```python
movie_data = "IMDB movies.csv"

movie_df = pd.read_csv(movie_data)
movie_df.head()
```

```python
company_df = movie_df[["production_company", ""]]
company_df
```

```python
company_df.dropna(how="any")
```

```python
filtered_df = pd.DataFrame(company_df[["production_company"]].value_counts())
filtered_df.sort_values(0, ascending=False)
```

```python
filtered_df.to_csv("production.csv", encoding="utf-8", index=True, header=True)
```
