```python
import pandas as pd
import numpy as np
```

```python
movie_data = "IMDB movies.csv"

movie_df = pd.read_csv(movie_data)
movie_df
```

```python
del movie_df["title"]
del movie_df["Unnamed: 22"]
del movie_df["Over budget"]
movie_df['country'] = movie_df['country'].str.split(',').str[0]
movie_df['genre'] = movie_df['genre'].str.split(',').str[0]
movie_df.head()
```

```python
filtered_df = movie_df.loc[movie_df["country"] == "USA"]
filtered_df = filtered_df.loc[filtered_df["year"] >= 2010]
filtered_df
```

```python
filtered_df = filtered_df[['genre', "year", "worlwide_gross_income"]]
filtered_df[filtered_df.columns[2:]] = filtered_df[filtered_df.columns[2:]].replace('[\$,]', '', regex=True).astype(float)
filtered_df.dropna()
```

```python
genre_filter = filtered_df[(filtered_df["genre"] == "Action") | (filtered_df["genre"] == "Comedy")
                          | (filtered_df["genre"] == "Adventure") | (filtered_df["genre"] == "Animation") | (filtered_df["genre"] == "Biography")
                          | (filtered_df["genre"] == "Crime")| (filtered_df["genre"] == "Drame") | (filtered_df["genre"] == "Sci-Fi")
                          | (filtered_df["genre"] == "Horror") | (filtered_df["genre"] == "Fantasy") | (filtered_df["genre"] == "Family")]
genre_filter = genre_filter.dropna()
genre_filter
```

```python
final_df = genre_filter.groupby(["genre", "year"])
final_df.count().head()
```

```python
print_df = pd.DataFrame(final_df["worlwide_gross_income"].sum())
```

```python
print_df.to_csv("genre.csv", encoding="utf-8", index=True, header=True)
```
