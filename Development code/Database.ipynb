```python
import pandas as pd
import numpy as np
```

```python
movie_data = "IMDB movies.csv"
movie_df = pd.read_csv(movie_data)
```

```python
del movie_df["title"]
del movie_df["Unnamed: 22"]
del movie_df["Over budget"]
movie_df['country'] = movie_df['country'].str.split(',').str[0]
movie_df['genre'] = movie_df['genre'].str.split(',').str[0]
movie_df.head()
```

```python
filtered_df = movie_df.loc[(movie_df["country"] == "USA") | (movie_df["country"] == "Australia") | (movie_df["country"] == "UK") 
                          | (movie_df["country"] == "New Zealand")]
filtered_df.head()
```

```python
final_df = filtered_df[['original_title', "year", "genre", "duration", "country", "director", "writer", "actors", "description"
                       , "avg_vote"]]
final_df
```

```python
final_df.to_csv("database.csv", encoding="utf-8", index=False, header=True)
```

```python
title_df = final_df["original_title"]
print_df = pd.DataFrame(title_df)
print_df 
```

```python
print_df.to_csv("titles.csv", encoding="utf-8", index=False, header=True)
```
