This file is transforming the main file into different csv to be used during the analysis.

```python
import pandas as pd
import numpy as np
```

```python
movie_data = "IMDB movies.csv"

movie_df = pd.read_csv(movie_data)
movie_df
```

```python
del movie_df["title"]
del movie_df["Unnamed: 22"]
del movie_df["Over budget"]
movie_df.head()
```

```python
movie_df['country'] = movie_df['country'].str.split(',').str[0]
movie_df.tail()
```

```python
movie_df['genre'] = movie_df['genre'].str.split(',').str[0]
movie_df.tail()
```

```python
first_movie_df = movie_df[["country", "original_title", "year", "genre","votes","avg_vote"]]
first_movie_df
```

```python
filtered_df = first_movie_df.loc[first_movie_df["votes"] >= 2000]
filtered_df = filtered_df.sort_values("avg_vote", ascending=False)
filtered_df = filtered_df.loc[filtered_df["country"] == "USA"]
filtered_df.head(10)
```

```python
filtered_df.to_csv("graph1-highest_rated.csv", encoding="utf-8", index=False, header=True)
```
