```python
import pandas as pd
import numpy as np
```

```python
movie_data = "IMDB movies.csv"
movie_df = pd.read_csv(movie_data)
del movie_df["title"]
del movie_df["Unnamed: 22"]
del movie_df["Over budget"]
movie_df['country'] = movie_df['country'].str.split(',').str[0]
movie_df['genre'] = movie_df['genre'].str.split(',').str[0]
```

```python
profit_budget_df = movie_df[["country", "original_title", "year", "genre","avg_vote", "votes", "budget", "worlwide_gross_income"]]
profit_budget_df 
```

```python
profit_budget_df = profit_budget_df.dropna(how='any')
profit_budget_df[profit_budget_df.columns[6:]] = profit_budget_df[profit_budget_df.columns[6:]].replace('[\$,]', '', regex=True).astype(float)
profit_budget_df
```

```python
profit_budget_df.to_csv("graph4-budget vs profit.csv", encoding="utf-8", index=False, header=True)
```
