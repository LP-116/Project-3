```python
import pandas as pd
import numpy as np
```

```python
movie_data = "IMDB movies.csv"
movie_df = pd.read_csv(movie_data)
del movie_df["title"]
del movie_df["Unnamed: 22"]
del movie_df["Over budget"]
movie_df.head()
movie_df['country'] = movie_df['country'].str.split(',').str[0]
movie_df['genre'] = movie_df['genre'].str.split(',').str[0]
```

```python
votes_df = movie_df[["country", "original_title", "year", "genre","avg_vote", "votes", "worlwide_gross_income"]]
votes_df 
```

```python
votes_df = votes_df.dropna(how='any').copy()
```

```python
votes_df[votes_df.columns[6:]] = votes_df[votes_df.columns[6:]].replace('[\$,]', '', regex=True).astype(float)
```

```python
votes_df.to_csv("votes_unfiltered(profit vs rating).csv", encoding="utf-8", index=False, header=True)
```

```python
votes_df = votes_df.loc[votes_df["votes"] >= 2000]
```
