```python
import pandas as pd
import numpy as np
```

```python
movie_data = "IMDB movies.csv"
movie_df = pd.read_csv(movie_data)
del movie_df["title"]
del movie_df["Unnamed: 22"]
del movie_df["Over budget"]
movie_df.head()
```

```python
movie_df['country'] = movie_df['country'].str.split(',').str[0]
movie_df['genre'] = movie_df['genre'].str.split(',').str[0]
```

```python
second_movie_df = movie_df[["country", "original_title", "year", "genre","avg_vote", "worlwide_gross_income"]]
second_movie_df
```

```python
second_movie_df = second_movie_df.dropna(how='any')
second_movie_df
```

```python
second_movie_df[second_movie_df.columns[5:]] = second_movie_df[second_movie_df.columns[5:]].replace('[\$,]', '', regex=True).astype(float)
```

```python
second_movie_df
```

```python
second_movie_df = second_movie_df.sort_values("worlwide_gross_income", ascending=False)
```

```python
second_movie_df.to_csv("graph2-profit.csv", encoding="utf-8", index=False, header=True)
```
